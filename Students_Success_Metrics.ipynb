#1.Select all records where the student's Exam Score is above 85.

```python
df[df["Exam_Score"]>85]
```

#2.Sort the DataFrame based on total marks in descending order.


```python
df.sort_values(by=["Exam_Score"],ascending=False)
```

#3.Rank the students based on their total marks and add the rank as a new column.


```python

df['Rank'] = df['Exam_Score'].rank(method='min', ascending=False).astype(int)
df_sorted = df.sort_values(by='Rank')
df_sorted
```

#4.The count of students who failed (score < 40) in exam scores

```python
df[df["Exam_Score"]<40].value_counts("Exam_Score")
```

#5.Display the exam score and count of female students study in public school without internet?

```python
df[(df["Gender"]=="Female") & (df["School_Type"]=="Public") & (df["Internet_Access"]=="No")]["Exam_Score"].value_counts()
```
